#### 5.3 Some extremely basic concepts of early Life Course Theory.

As mentioned above, the tenents of the FFC are grounded in the basics of Life Course Research, and specifically, in the area of 'Transitions to Adulthood'. That is beacuse, after all, we're predicting what happens at age 15 of a child's life, given information on the child up and until age 9.

'Traditions to Adulthood' are frequently discussed within sociological literature, but what are they?

Examples include; completion of education, leaving parental home, entry into the labour market, entry into partnerships,  entry into parenthood.

The concept of a generation/path dependence thesis is by now well established in the academic literature (Mannheim 1952; Inglehar 1977; Becker 1991, etc):
* Life course patterns and pathways vary from generation to generation.
* According to `birth cohorts', which place individuals specific cultural resources and combinations of constraints and opportunities.
* Size of cohorts also matters (i.e. Ryder, 1965)
* Persistent inequalities and stratification: by social class and gender.


Importantly, the *'Individualization Thesis'* (Beck 1986; Hohli 1986; Giddens 1990; Wallace & Kovatcheva 1998) runs alongside ideas of path dependence. The idea concerns the 'Destabilization of the Life Course: there is no such thing as a 'normal' biography any more; everybody increasingly lives 'fragmented' and 'individualized' trajectories, and there are more reversible life events. There are no longer -- or a least there is now much less age-specificity when it comes to when paths or events occurs. There is a decreasing influence and significance in social class and gender (although huge inequalities still persist). This idea of the individualisation hypothesis is also a key component of the Second Demographic Transition theory (which entails “sustained sub-replacement fertility, a multitude of living arrangements other than marriage, the disconnection between marriage and procreation, and no stationary population"). Trends towards de-standardisation (or rather, "new standardization") across countries is called the "convergence thesis".

What is the _role of the state_ throught the lifecourse? Our lives are shaped widely the institutionalisation of life courses through the state. This takes the form of `Collective functionalism'; starting from participating in educational institutions, to the provision of pensions at old age. Furthermore, the state regulates norms (e.g. age to drive, or buy cigarettes), and age-stratifies through institutions (e.g. the typical age for enrollment in university, and therefore first employment), policies (age to marry), laws (legal age at adulthood), and regulations (compulsory military service). Mayer & Schoepflin famously wrote in 1989 about how the possibility of the American welfare state was potentially weakening social bonds and family ties, concluding that it also led to more individualistic life course trajectories across nations and time.

Life Course Research has generally been shaped enormously by the rise of large longitudinal datasets. Beginning in the 1940s/1950s, they have become what's known as the 'Hubble Telescope' of the Social Sciences. Some prominent examples which you might see include the Health and Retirement Survey, the UK Household Longitudinal Survey, and so forth. Some focus on older survey participants (like ELSA), and some focus more on the transitions to adulthood (like the FFCWS). Some collect DNA, some follow entire family lineages.

#### 5.4 Introducing the FFC Data

Before we talk specifically about the dataset, we should learn and understand more about how the ‘Common Task Method’ was used. An organizer (in this case, Professor Salganik) designs a prediction task and recruits researchers.

Of critical importance: The exact same outcomes using the exact same data. Predictions are evaluated with the exact same error metric on held-out data. The held-out data is\was held by the organizer and not available to participants. The participants are free to use any technique to generate predictions. The standardization of the prediction task facilitates comparisons between different approaches. It also removes concerns about over-fitting, researcher degrees of freedom. In this case -- as opposed to many others within the social sicences -- a Common Task Method was feasible because the heldout data was _truly_ unseen; Professor Salganik entered into an agreement with the data provider to not realise Wave Six of their large longitudinal survey until the 'Challenge' was complete. This is critical, and meant that the 'test' data really was unseen.

The [Fragile Families and Child Wellbeing Study](https://fragilefamilies.princeton.edu/) (henceforth known as the FFCWS) is a multistage, stratified random sample. It has, to date, been used  more than 750 published journal articles, where the majority of those works are inferential in nature. It contains extrmely rich longitudinal data on thousands of families, with a 3:1 oversample of births to non-married parents. Each mother gave birth to a child in a large (200k+) US city (of which 20 were included, including two cities used in the pilot of the study). The study began in the year 2000. The original motivation was to understand families formed by unmarried parents, and importantly, the lives of children born into these families. The study had six waves: child birth (Wave 1) and ages 1 (Wave 2), 3 (Wave 3), 5 (Wave 4), 9 (Wave 5), and 15 (Wave 6). Each wave includes a number of different data collection modules, and responses from various people involved in the life of the child.

Each data collection module is made up of around 10 sections. Different family members/associates are interviewed at different times. In-person tests conducted in waves 3-5 (i.e. Peabody vocabulary test), as
well as measurements, home inspections, and so forth. Over time, the scope of data collection increased significantly.

#### 5.4.1 Introducing the Evaluation Metric

Models were evaluated with the R Holdout metric:

$$ R^2 = 1-\frac{\sum_{i\in \textrm{Holdout}} (y_i - \hat{y}_i)^2}{\sum_{i \in \textrm{Holdout}} (y_i - \bar{y}_{\textrm{Training}})^2}$$

A few very important things to note here. This metric is used for both the set of three continuous variables,and the set of three binary variables. That itself implies that participants were asked to submit probabalistic evaluations for the binary outcomes (otherwise the $R^2$ metric would give a very curious result). Importantly, note that an $R^2$ Holdout equal to 0 is no more accurate than predicting the mean of the training data. Note also, that there is no lower bound! This measure can be infinitely negative! In contrast, an R Holdout of 1 predicts perfectly.

#### 5.4.5.1 Unpredictability

Some ([very famous](https://www.filizgarip.com/)) commentators interpretted the results of the Fragile Families Challenge to mean that life outcomes were not predictable. Importantly, see [this](https://www.pnas.org/content/117/15/8234.short) commentary published syncronously with the main PNAS article itself. But is that necessarily the case?

Some _socio-economic_ reasons as to why perhaps it is not:

* Three aspects of timing may decrease predictability:
    1. Six year gap between waves 5 and 6.
    2. A large social disruption (the Great Recession).
    3. The prediction of wave 6 comes at a time when the child was 15 years old.

* This may be a particularly turbulent time for children and families.
* A largely urban and disadvantaged group living in the contemporary United
States may have more unpredictable lives than other groups.
* The results would have been qualitatively different if we selected different outcomes from wave 6 (age 15).

Some _data-scientific_ reasons as to why perhaps it is not:

Four (or Five?) reasons to be skeptical:

1. The curse of dimensionality: With just 4,242 families, there isnt that much space for ML to stretch its legs!

2. Researchers ‘had access to thousands of predictor variables’: did they?

3. Of the 12,942 variables, 2,358 were constant. As shown above, 'Missing-ness' is substantial.
4. How homogeneous was the community? Mostly, social data scientists, as opposed to deep learning engineers.
5. How appropriate is the R Holdout metric?

Or does it mean – as Garip (2020, PNAS) puts it:

_“...that life outcomes are too idiosyncratic and subject to a predictability ceiling.”_
